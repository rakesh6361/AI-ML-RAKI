# Filtering



```python
# 1)

def sal(a):
    if a>5000:
        return True
    
l1=[2000,4000,2500,6000,10000,4900]
f=list(filter(sal,l1))
f
```

```python
# 2)
def de(a):
    if a<1000:
        return True
    
l1=[1500,900,6000,549,2300,500,2390,345,678]
res=list(filter(de,l1))
res
```

# Reducing



```python
# 1) 

from functools import reduce
def add(x,y):
    return x+y
l1=[10,20,30,40,50]
res=reduce(add,l1)
res
```

```python
# 2)

from functools import reduce
def sqr(a,b):
    return a+10
l1=[2,3,4,5,6]
res=reduce(sqr,l1)
res
```
