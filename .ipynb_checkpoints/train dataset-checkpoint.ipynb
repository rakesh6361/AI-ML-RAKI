#### 5. Mean, Mode, Median

```python

df['account_encoded'] = df['account.type'].map({'human': 0, 'bot': 1})

df['account_encoded'].mean()
df['account_encoded'].median()
df['account_encoded'].mode()[0]
```

#### 6. Agg

```python
df.groupby('account.type').agg({
    'account_encoded': ['mean', 'std', 'sum']
})
```

#### 7. Std

```python
df['account_encoded'].std()
```

```python
df['account_encoded'].sum()
```

#### 9. Vectorization

```python
df['account_encoded_squared'] = df['account_encoded'] **2
df
```

#### 10. Map

```python
df['name_length'] = df['screen_name'].map(len)
df
```

#### 12. Reduce

```python
from functools import reduce
sample_list = [1, 2, 3, 4]
reduce(lambda x, y: x + y, sample_list)
```

#### 13. Lambda

```python
df['short_name'] = df['screen_name'].apply(lambda x: x.split('#')[0])
df
```
